```python
m = Basemap(projection='hammer', lon_0=-90)

# convierte latitudes y longitudes en la proyeccion geolocal
x, y = m(lons, lats)

m.drawmapboundary(fill_color='#99FFFF')
m.fillcontinents(color='#CC9966', lake_color='#99FFFF')
m.scatter(x,y,24,marker='o',color='k')
plt.text(x[-1],y[-1],'LANIA',color='b')
```

```python
# obtener el timestamp del documento
timestamp = sensorData.iloc[:,32:33]
```

```python
# ver el contenido de los valores de timestamp
timestamp.values
```

```python
# mostrar los primeros 5 registros
timestamp[:5]
```

```python
from datetime import datetime
import time
import calendar

#separar la fecha de la hora de un valor dell timestamp leido con el sensor
momento = str(timestamp.values[-1]).split()


# --- FECHA ---
fecha = momento[0]
fechaCortada = fecha.split("-")

# obtener el dia, mes y año
aa = int(fechaCortada[0][2:]) #comenzar en la segunda posicion del string, para obtener el año
mm = int(fechaCortada[1])
dd = int(fechaCortada[2])


# --- HORA ---
hora = momento[1]
horaCortada = hora.split(":")

# obtener la hora, minuto y sgeundo
hr = int(horaCortada[0])
mn = int(horaCortada[1])
sg = int(horaCortada[2])

# --- DEFINIR OBJETO DE FECHA
objetoFecha = datetime(aa, mm, dd, hr, mn, sg)
```

```python
# ver el contenido del objeto de fecha
objetoFecha
```

```python
# convertir el objeto fecha a un formato de calendario
calendario = calendar.timegm(objetoFecha.timetuple())
```

```python
map = Basemap(projection='mill',lon_0=-90)
# plot coastlines, draw label meridians and parallels.
map.drawcoastlines()
map.drawparallels(np.arange(-90,90,30),labels=[1,0,0,0])
map.drawmeridians(np.arange(map.lonmin,map.lonmax+30,60),labels=[0,0,0,1])
# fill continents 'coral' (with zorder=0), color wet areas 'aqua'
map.drawmapboundary(fill_color='#99ffff')
map.fillcontinents(color='coral',lake_color='#99ffff')
x, y = map(lons,lats)
map.scatter(x,y,24,marker='o', color='b')
plt.text(x[-1],y[-1],'LANIA',color='b');

# agregar sombreado UTC
date=datetime.utcfromtimestamp(calendario)
CS=map.nightshade(date)
plt.title('Day/Night Map for %s (UTC)' % date.strftime("%d %b %Y %H:%M:%S"));
```

```python
# Acercar mas el mapa

map = Basemap(projection='mill',lon_0=-90, lat_0=0,
    llcrnrlon=-120, llcrnrlat=0.0,
    urcrnrlon=-70, urcrnrlat=30.75)
# plot coastlines, draw label meridians and parallels.
map.drawcoastlines()
map.drawparallels(np.arange(map.latmin,map.latmax,10),labels=[1,1,1,1])
map.drawmeridians(np.arange(map.lonmin,map.lonmax,20),labels=[0,0,0,1])
# fill continents 'coral' (with zorder=0), color wet areas 'aqua'a
map.drawmapboundary(fill_color='#99ffff')
map.fillcontinents(color='coral',lake_color='#99ffff')
x, y = map(lons,lats)

map.plot(x, y, 'bo', markersize=18)
plt.text(x[-1],y[-1],'LANIA');
```
